```python
import os as os
import pandas as pd
from itertools import product
import numpy as np
import scipy.stats as ss
```

```python
df = pd.read_csv(r"data/train.csv")
```

```python
cat_cols = ["HomePlanet", "CryoSleep","Destination","VIP",'Transported'] #leave out Cabin
```

```python
df = df[cat_cols].copy()
```

```python
df.dropna(inplace=True)
```

```python
df_dummy = pd.get_dummies(df).copy()
```

```python
df_dummy.dropna(inplace=True)
```

```python
cat_1_list = list(df_dummy.columns)
cat_2_list = cat_1_list.copy()
```

```python
cat_var_prod = list(product(cat_1_list,cat_2_list, repeat = 1))
```

```python
result = []
for i in cat_var_prod:
    if i[0] != i[1]:
        result.append((i[0],i[1],list(ss.chi2_contingency(pd.crosstab(
                                df_dummy[i[0]], df_dummy[i[1]])))[1]))
```

```python
chi_test_output = pd.DataFrame(result, columns = ['var1', 'var2', 
                                                       'coeff'])
df_cats = chi_test_output.pivot(index='var1', columns='var2', values='coeff').copy()
```

```python
df_cats[df_cats["Transported"]<=0.05]
```
